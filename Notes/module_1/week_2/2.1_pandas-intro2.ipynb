==> Confusing!

When does pandas know when I want to subset columns or when I want to subset rows??


-> use `.loc[]` or `iloc[]`

Example:
```
df.loc[rows, columns]
```

````python
df.loc[df["orange_density"].lt(.5), ["orange_mass"]]
````

````python
df.loc[df["kiwi_mass"].gt(.3),]
````

````python
df.loc[:,["kiwi_density"]]
````

````python
df.loc[:,["kiwi_mass", "orange_density"]]
````

```
df.iloc[]
```

allows you to use positional indexes instead of labels

````python
df.index = range(10, 20 ,1)
````

````python
df
````

````python
df.loc[1:4,]
````

`.loc` is not returning anything because it is looking for index labels 1-4

````python
df.iloc[1:4]
````

````python
df.iloc[1:7, 2:4]
````
